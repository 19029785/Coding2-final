```python
import random

print("Welcome to guessing number game！")
print("I've thought of a number between 1 and 100, and you need to guess it in the fewest number of times.")

number = random.randint(1, 100)
guesses = 0

while True:
    guess = int(input("Please guess a number:"))
    guesses += 1
    
    if guess < number:
        print("Too small, please try again.")
    elif guess > number:
        print("Too big, please try again.")
    else:
        print(f"Congratulations! You guessed it right! the number is {number}，you used {guesses} times.")
        break
```

```python
import random

print("Welcome to the game of rock paper scissors！")
print("You need to play a best-of-three match against the computer to see who gets two wins first. Please remember to capitalize the first letter.")

options = ["Rock", "Scissors", "Paper"]
win_rules = {"Rock": "Scissors", "Scissors": "Paper", "Paper": "Rock"}
player_wins = 0
computer_wins = 0

while player_wins < 2 and computer_wins < 2:
    print(f"\n current score：You {player_wins} - {computer_wins} computer")
    player_choice = input("Choose please (rock/scissors/paper)：")
    
    if player_choice not in options:
        print("Invalid move, please select again.")
        continue
    
    computer_choice = random.choice(options)
    print(f"Computer's turn：{computer_choice}")
    
    if player_choice == computer_choice:
        print("Draw！")
    elif win_rules[player_choice] == computer_choice:
        print("You win！")
        player_wins += 1
    else:
        print("Computer wins！")
        computer_wins += 1

print("\n Game over！")
if player_wins > computer_wins:
    print("Congratulations. You won the game！")
else:
    print("Unfortunately, the computer won the game.")
```

```python
import random
import time

# Defines where the mole appear
holes = ['  O    O    O  ',
         '    O O O     ',
         '  O O O O O   ',
         '  O   O   O   ',
         '    O   O     ']

# game rule tip
print("Whack-a-mole game, press any letters to start.")
input()

# game start
score = 0
while True:
    #randomly pick a position
    hole = random.choice(holes)
    
    # output the location where the mole appears
    print('\n\n\n')
    print('    /\_/\ ')
    print('   / o o \\')
    print('  (   "   )')
    print(hole)
    print('  (       )')
    print('   \~~~~~/')
    print('    `---` ')
    
    # record time
    appear_time = time.time()
    
    # wait player hit the mole
    input("Hit enter when you're ready!")
    
    # Determine if the player hits the gopher
    if time.time() - appear_time > 1.5:
        print("It's been too long and the mole have escaped!")
    else:
        print("Congratulations, you hit the mole！")
        score += 1
    
    # score
    print("current score：", score)
    
    # game again or not
    choice = input("Do you want to continue the game？（Y/N）").lower().strip()
    if choice == 'n':
        break
```

```python
import random

# letters
letters = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L",]

# 4*4 form
matrix = []
for i in range(4):
    row = []
    for j in range(4):
        row.append(random.choice(letters))
    matrix.append(row)


for row in matrix:
    print(row)

# game rule tips
print("A letter link game, eliminate all adjacent identical letters to win.")
print("The format of coordinates you should use is： x,y")
print("Tip：x represents the number in which line, y represents the order of letter in this line. x and y start at 0.")


# game start
while True:
    # get player's input
    x1, y1 = map(int, input("Please enter the coordinates of the first letter(x,y)：").split(","))
    x2, y2 = map(int, input("Please enter the coordinates of the second letter(x,y)：").split(","))
    
    # Determine whether the input is valid
    if x1 < 0 or x1 > 3 or y1 < 0 or y1 > 3 or x2 < 0 or x2 > 3 or y2 < 0 or y2 > 3:
        print("Invalid input, please re-enter!")
        continue
    
    # Check if two letters are the same
    if matrix[x1][y1] != matrix[x2][y2]:
        print("The two letters are not the same, please re-enter!")
        continue
    
    # Check if two letters are adjacent
    if abs(x1 - x2) + abs(y1 - y2) != 1:
        print("Two letters are not adjacent, please re-enter!")
        continue
    
    # eliminate two letters
    matrix[x1][y1] = " "
    matrix[x2][y2] = " "
    
    # output the current matrix
    for row in matrix:
        print(row)
    
    # Determine if the game is over
    if all([letter == " " for row in matrix for letter in row]):
        print("Congratulations, you have won!")
        break
```

```python
from PIL import Image, ImageFilter

# image
image = Image.open("image.jpg")

# filter
def apply_filter(filter_name):
    if filter_name == "BLUR":
        image_filtered = image.filter(ImageFilter.BLUR)
    elif filter_name == "CONTOUR":
        image_filtered = image.filter(ImageFilter.CONTOUR)
    elif filter_name == "DETAIL":
        image_filtered = image.filter(ImageFilter.DETAIL)
    elif filter_name == "EDGE_ENHANCE":
        image_filtered = image.filter(ImageFilter.EDGE_ENHANCE)
    elif filter_name == "EMBOSS":
        image_filtered = image.filter(ImageFilter.EMBOSS)
    elif filter_name == "FIND_EDGES":
        image_filtered = image.filter(ImageFilter.FIND_EDGES)
    elif filter_name == "SHARPEN":
        image_filtered = image.filter(ImageFilter.SHARPEN)
    else:
        image_filtered = image
    
    # image after filter
    image_filtered.show()

# game
while True:
    # choose one filter
    filter_name = input("Please enter a filter name（BLUR, CONTOUR, DETAIL, EDGE_ENHANCE, EMBOSS, FIND_EDGES, SHARPEN），enter quit to stop the game：").upper()
    if filter_name == "QUIT":
        break
    
    # use filter
    apply_filter(filter_name)
```

```python
import tkinter as tk


class PixelArtApp:
    def __init__(self, master, width=640, height=480, pixel_size=10, num_x=None, num_y=None):
        self.master = master
        self.master.geometry(f"{width}x{height}")
        self.master.title("pixel painting game")

        self.canvas = tk.Canvas(master, width=width, height=height, bg="white")
        self.canvas.pack()

        self.pixel_size = pixel_size
        self.num_x = num_x or width // pixel_size
        self.num_y = num_y or height // pixel_size
        self.colors = ["black", "red", "green", "blue"]  # set colors
        self.default_color_index = 0

        self.pixels = [[self.default_color_index for _ in range(self.num_x)] for _ in range(self.num_y)]

        self.canvas.bind("<Button-1>", self.handle_click)
        self.canvas.bind("<Button-3>", self.clear_canvas)
        self.master.bind("<Key>", self.handle_key)

    def draw_pixel(self, x, y, color):
        x1, y1 = x * self.pixel_size, y * self.pixel_size
        x2, y2 = x1 + self.pixel_size, y1 + self.pixel_size
        self.canvas.create_rectangle(x1, y1, x2, y2, fill=color, outline="white")

    def handle_click(self, event):
        x, y = event.x // self.pixel_size, event.y // self.pixel_size
        if x < self.num_x and y < self.num_y:
            self.pixels[y][x] = (self.pixels[y][x] + 1) % len(self.colors)
            color = self.colors[self.pixels[y][x]]
            self.draw_pixel(x, y, color)

    def clear_canvas(self, event):
        self.pixels = [[self.default_color_index for _ in range(self.num_x)] for _ in range(self.num_y)]
        self.canvas.delete("all")

    def handle_key(self, event):
        if event.char.isdigit():
            index = int(event.char) - 1
            if index < len(self.colors):
                self.default_color_index = index

    def run(self):
        self.master.mainloop()


if __name__ == "__main__":
    app = PixelArtApp(tk.Tk())
    app.run()
```

```python
import tkinter as tk
from PIL import Image, ImageTk

class PuzzleGame:
    def __init__(self, master):
        self.master = master
        self.master.title('Puzzle Game')
        self.images = [Image.open('image1.png'),
                       Image.open('image2.png'),
                       Image.open('image3.png'),
                       Image.open('image4.png')]
    
        self.rotations = [0, 90, 180, 270]
        self.current_positions = [0, 1, 2, 3]
        self.current_image = self.get_current_image()
        self.create_widgets()

    def create_widgets(self):
        # Create a canvas and show the current image
        self.canvas = tk.Canvas(self.master, width=600, height=750)
        self.canvas.pack()
        self.canvas.create_image(0, 0, anchor='nw', image=self.current_image)

        # Create buttons for rotating each image
        self.rotate_buttons = []
        for i in range(4):
            button = tk.Button(self.master, text=f'Rotate {i+1}', command=lambda idx=i: self.rotate_image(idx))
            button.pack(side=tk.LEFT, padx=10)
            self.rotate_buttons.append(button)

    def get_current_image(self):
        # Get the current positions of the images and their rotations
        current_images = [self.images[i].rotate(self.rotations[self.current_positions[i]]) for i in range(4)]
        

# Resize each image to fit the canvas
        max_width = 300
        max_height = 375
        resized_images = [image.resize((max_width, max_height)) for image in current_images]

        
        # Combine the images into a single image and convert to Tkinter PhotoImage
        combined_image = Image.new('RGB', (max_width*2, max_height*2))
        combined_image.paste(resized_images[0], (0, 0))
        combined_image.paste(resized_images[1], (max_width, 0))
        combined_image.paste(resized_images[2], (0, max_height))
        combined_image.paste(resized_images[3], (max_width, max_height))
        return ImageTk.PhotoImage(combined_image)

    def rotate_image(self, idx):
        # Rotate the image at the given index
        self.current_positions[idx] = (self.current_positions[idx] + 1) % 4
        
        # Update the current image and redraw it on the canvas
        self.current_image = self.get_current_image()
        self.canvas.delete('all')
        self.canvas.create_image(0, 0, anchor='nw', image=self.current_image)

if __name__ == '__main__':
    root = tk.Tk()
    app = PuzzleGame(root)
    root.mainloop()
```

```python
import pygame
import random
import math

pygame.init()

# set window
screen_width = 1000
screen_height = 800
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Boids")


NUM_BOIDS = 50
BOID_SIZE =3
BOID_SPEED = 5
BOID_TURN_RATE = 5


class Boid:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.angle = random.randint(0, 360)
        self.speed = BOID_SPEED

    # birds location
    def update(self):
        self.x += self.speed * math.cos(math.radians(self.angle))
        self.y += self.speed * math.sin(math.radians(self.angle))

    # draw birds
    def draw(self):
        pygame.draw.circle(screen, (0, 255, 255), (int(self.x), int(self.y)), BOID_SIZE)

    # ditance
    def distance(self, other_boid):
        dx = self.x - other_boid.x
        dy = self.y - other_boid.y
        return math.sqrt(dx**2 + dy**2)

    # direction
    def get_direction_to_mouse(self, mouse_pos):
        dx = mouse_pos[0] - self.x
        dy = mouse_pos[1] - self.y
        angle_to_mouse = math.atan2(dy, dx)
        return math.degrees(angle_to_mouse)

   
    def get_desired_angle(self, mouse_pos):
        angle_to_mouse = self.get_direction_to_mouse(mouse_pos)
        return (angle_to_mouse + self.angle) / 2

    # refresh birds direction
    def update_angle(self, other_boids):
        sum_angles = 0
        num_neighbors = 0
        for boid in other_boids:
            if self.distance(boid) < 50 and boid != self:
                sum_angles += boid.angle
                num_neighbors += 1
        if num_neighbors > 0:
            avg_angle = sum_angles / num_neighbors
            if abs(avg_angle - self.angle) > BOID_TURN_RATE:
                if (avg_angle - self.angle + 360) % 360 < 180:
                    self.angle += BOID_TURN_RATE
                else:
                    self.angle -= BOID_TURN_RATE

    # draw new direction birds
    def simulate(self, boids, mouse_pos):
        for boid in boids:
            boid.update_angle(boids)
        for boid in boids:
            desired_angle = boid.get_desired_angle(mouse_pos)
            if abs(desired_angle - boid.angle) > BOID_TURN_RATE:
                if (desired_angle - boid.angle + 360) % 360 < 180:
                    boid.angle += BOID_TURN_RATE
                else:
                    boid.angle -= BOID_TURN_RATE
            boid.update()
            boid.draw()


boids = [Boid(random.randint(0, screen_width), random.randint(0, screen_height)) for _ in range(NUM_BOIDS)]

# loop
running = True
while running:

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # mouse location
    mouse_pos = pygame.mouse.get_pos()

    screen.fill((0, 0, 0))
    for boid in boids:
        boid.simulate(boids, mouse_pos)
    pygame.display.flip()

# quit Pygame
pygame.quit()
```

```python
import pygame
import random

pygame.init()


window_size = (1200, 1000)
screen = pygame.display.set_mode(window_size)
pygame.display.set_caption("Music and Image Game")

# images
background_image = pygame.image.load("background.png")
background_image = pygame.transform.scale(background_image, (int(background_image.get_width() * 0.4), int(background_image.get_height() * 0.4)))

images = [pygame.image.load("image5.png"), pygame.image.load("image6.png"), pygame.image.load("image7.png")]
for i in range(len(images)):
    images[i] = pygame.transform.scale(images[i], (int(images[i].get_width() * 0.2), int(images[i].get_height() * 0.2)))
font = pygame.font.Font(None, 36)
music_file = "seventeen.mp3"


image_position = [400, 300]

# play music
pygame.mixer.music.load(music_file)
pygame.mixer.music.play()

# music
music_info = pygame.mixer.Sound(music_file)
music_length = music_info.get_length()
music_bpm = 120
music_beat_duration = 60 / music_bpm
music_beat_count = int(music_length / music_beat_duration)

# loop
running = True
beat_count = 0
current_image = None
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # game situation
    current_time = pygame.mixer.music.get_pos() / 1000
    current_beat = int(current_time / music_beat_duration)
    if current_beat > beat_count:
        beat_count = current_beat
        current_image = random.choice(images)
        image_position = [random.randint(0, window_size[0] - current_image.get_width()), random.randint(0, window_size[1] - current_image.get_height())]

   
    screen.blit(background_image, [0, 0])
    if current_image:
        screen.blit(current_image, image_position)
    text = font.render("Current Beat: " + str(current_beat) + "/" + str(music_beat_count), True, (255, 255, 255))
    screen.blit(text, [10, 10])

    pygame.display.update()

pygame.quit()
```
